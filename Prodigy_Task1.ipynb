```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("/content/titanic_train.csv")
data.head(5)
```

```python
data.info()
```

Since the above data does not contain null values we move on with further analysis


```python
data['Age'].plot.hist(bins=80,color='yellow',edgecolor='black')
```

The above graph shows the distribution of age among the passangers.We can infer that majority of the passangers fall within the range of 18-40 years.

```python
sns.countplot (x='Sex',data=data)
```

The proportion of males compared to females is higher

```python
sns.countplot(x='Survived',hue='Sex',data=data)
```

We can see that number of males who did not survived is much higher than the number of females

```python
sns.countplot(x='Survived',hue='Pclass',data=data)
```

As per the above graph people belonging to the third class tend to have not survived.
